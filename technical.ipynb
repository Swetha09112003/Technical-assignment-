```python
import requests
import csv
url="https://jsonplaceholder.typicode.com/users"
response=requests.get(url)
users=response.json()
```

```python
filtered_users=[]
for user in users:
    if user["city"]=="Lebsackbury":
        filtered_users.append([user["id"],user["name"],user["email"],user["city"]])
        
```

```python
with open("filtered_users.csv","w",newline="") as file:
    writer =csv.writer(file)
    writer.writerow(["id","name","email","city"])
    writer.writerows(filtered_users)
print("done@ check filtered_users.csv file")
```
